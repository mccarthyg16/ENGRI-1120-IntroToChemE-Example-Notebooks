### ENGRI 1120: Dynamic Solution of a Single Mass Action Reaction

```julia
import Pkg; Pkg.activate("."); Pkg.resolve(); Pkg.instantiate();
```

```julia
using Plots
using Colors
using DifferentialEquations
```

```julia
include("ENGRI-1120-Example-CodeLib.jl");
```

```julia
# setup the stoichiometric matrix -
S = Array{Float64,2}(undef, 3,1);
S[1,1] = -1.0;
S[2,1] = -1.0;
S[3,1] = 1.0;
```

```julia
# setup the initial condition vector -
initial_condition_vector = [2.0, 1.0, 0.0];
```

```julia
# setup rate constant vector -
k = Array{Float64,2}(undef,1,2)
k[1,1] = 1.0;
k[1,2] = 0.1;
```

```julia
# setup -
model_dictionary = Dict{String,Any}();
model_dictionary["S"] = S;
model_dictionary["k"] = k;
model_dictionary["initial_condition_vector"] = initial_condition_vector;
model_dictionary["number_of_dynamic_states"] = 3

# solve the model -
(T,X) = evaluate(model_dictionary; tspan=(0.0, 10.0));
```

```julia
# plot -
plot(T,X[:,1], label="A (rev)", lw=3, c=colorant"#89CCE2", bg="floralwhite", background_color_outside="white", framestyle = :box, fg_legend = :transparent);
plot!(T,X[:,2], label="B (rev)", lw=3, c=colorant"#EF4035");
plot!(T,X[:,3], label="C (rev)", lw=3, c=colorant"#6EB43F", ls=:dash);
xlabel!("Time (AU)", fontsize=20)
ylabel!("Concentration (AU)", fontsize=20)

#savefig("./figs/Fig-MAK-Example.pdf")
```
